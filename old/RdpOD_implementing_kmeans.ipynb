```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
import random
import bisect
from collections import OrderedDict
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv(r'D:\study\codes\engineering_codes\research\data_sets\statlog+landsat+satellite\taining_data_set.csv')
print(df)
```

```python
#This step indicates conversion of df from csv file to dataframe
df = pd.DataFrame(df)
```

```python
# checking datatypes using info method
df.info()
```

```python
# applying pca for the dataset
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(df)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])


print("Original data shape:", df.shape)
print("Transformed data shape:", principalComponents.shape)
print("Transformed data:")
print(principalDf)
```
